### This visualization shows the relationship between the severity of car accidents and whether they occurred during daylight or nighttime hours. The plot displays the count of accidents for each severity level, broken down by whether they occurred during sunrise/sunset or not.

```python
# Plot relationship between Severity and Daylight / Night
sns.countplot(y="Severity", hue="Sunrise_Sunset", data=df, orient='h')
plt.title("Relationship between Severity and Daylight / Night")
plt.xlabel("Count")
plt.ylabel("Severity")
plt.show()
```

### This code prepares the dataset for a visualization that shows the relationship between the severity of car accidents and whether they occurred on a weekday or weekend. We start by converting the 'Start_Time' column to datetime format and extracting the weekday name from it. We then create a new column called 'Time_of_Day' and replace 'Day' with 'Daylight' and 'Night' with 'Nighttime' for consistency. Finally, we create a new column called 'Weekday_Type' to represent whether the accident occurred on a weekday or weekend.

### The visualization displays the count of accidents for each severity level broken down by whether they occurred on a weekday or weekend. 

```python
# convert Start_Time to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# extract the weekday name from the Start_Time column
df['Weekday'] = df['Start_Time'].dt.day_name()

# create a new column 'Time_of_Day'
df['Time_of_Day'] = df['Sunrise_Sunset']

# replace 'Day' with 'Daylight' and 'Night' with 'Nighttime'
df['Time_of_Day'].replace({'Day': 'Daylight', 'Night': 'Nighttime'}, inplace=True)

# create a new column to represent whether the accident happened on a weekday or weekend
df['Weekday_Type'] = df['Weekday'].apply(lambda x: 'Weekday' if x in ['Monday','Tuesday','Wednesday','Thursday','Friday'] else 'Weekend')


# Plot relationship between Severity and Daylight / Night
sns.countplot(y="Severity", hue="Weekday_Type", data=df, orient='h')
plt.title("Relationship between Severity and Weekday / Week")
plt.xlabel("Count")
plt.ylabel("Severity")
plt.show()
```

### This code prepares the dataset for a visualization that shows the number of car accidents that occurred during each hour of the day. We start by creating a new column called 'Hour' that extracts the hour of the day from the 'Start_Time' column. We then group the data by the hour and count the number of accidents that occurred during each hour. This creates a new dataframe called 'hourly_counts'.

### This information can help you understand when accidents are most likely to occur and plan accordingly.

```python
# create a new column with the hour of the day
df['Hour'] = df['Start_Time'].dt.hour

# group the data by hour and count the number of accidents
hourly_counts = df.groupby('Hour')['ID'].count().reset_index()

# plot the data
sns.lineplot(x='Hour', y='ID', data=hourly_counts)
plt.title('Number of Accidents per Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

### Here, the code creates a new column in the dataframe to store the month of each accident using the dt.month_name() function. 

### The number of accidents is then counted for each month using the groupby() function and stored in the monthly_counts dataframe. 

###  Finally, the data is plotted using a line plot to visualize the trend in the number of accidents per month. This plot can help the user understand the seasonality of accidents throughout the year.

```python
# Create a new column with the month of the year
df['Month'] = df['Start_Time'].dt.month_name()

# Group the data by month and count the number of accidents for each month
monthly_counts = df.groupby('Month')['ID'].count().reset_index()

# Extract the first three letters of each month name
monthly_counts['Month'] = monthly_counts['Month'].str.slice(stop=3)

# Plot the data
sns.lineplot(x='Month', y='ID', data=monthly_counts)
plt.title('Number of Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
# Group the data by month and count the number of accidents for each month
monthly_counts = df.groupby('Month')['ID'].count().reset_index()

# Convert the month number to month name abbreviation
monthly_counts['Month'] = pd.to_datetime(monthly_counts['Month'], format='%B').dt.strftime('%b')

# Set the order of the x-axis categories
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_counts['Month'] = pd.Categorical(monthly_counts['Month'], categories=month_order, ordered=True)

# Plot the data
sns.lineplot(x='Month', y='ID', data=monthly_counts)
plt.title('Number of Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```

###  Finally, a heatmap is plotted using the heatmap() function from the Seaborn library to visualize the clustered bar chart. The heatmap is color-coded to indicate the severity of the accident and also displays the count of accidents for each combination of severity and weather condition as annotations.

### The code selects the columns of interest from the dataframe df and groups the data by Severity and Weather_Condition. Then it pivots the data to create a clustered bar chart and plots the chart using a heatmap with annotations. The resulting visualization shows the relationship between different weather conditions and accident severity.

```python
# select columns of interest
df1 = df[['Severity', 'Weather_Condition']]

# group the data by severity and weather condition
grouped_df = df1.groupby(['Weather_Condition', 'Severity']).size().reset_index(name='Count')

# pivot the data to create a clustered bar chart
pivot_df = grouped_df.pivot(index='Weather_Condition', columns='Severity', values='Count')

# plot the clustered bar chart
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, cmap='YlOrRd', annot=True, fmt='g')
plt.title('Accidents by Weather Condition and Severity')
plt.xlabel('Severity')
plt.ylabel('Weather Condition')
plt.show()
```

```python
# select columns of interest
df2 = df[['Severity', 'Weather_Condition']]

# filter the data for severity level 2
df2 = df[df['Severity'] == 2]

# group the data by weather condition
grouped_df = df2.groupby(['Weather_Condition']).size().reset_index(name='Count')

# plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(grouped_df.pivot(index='Weather_Condition', columns='Count', values='Count'), cmap='YlOrRd', annot=True, fmt='g')
plt.title('Accidents by Weather Condition (Severity 2)')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.show()
```

### The code groups the accident data by weather condition and counts the number of accidents for each condition. It then sets the minimum and maximum counts for scaling the color map of the heatmap that will be created using the Seaborn library. 

### Finally, it creates the heatmap using Seaborn with the previously calculated minimum and maximum counts, as well as a logarithmic color scale and a label for the color bar. The resulting heatmap allows for easy identification of the weather conditions that pose the highest risk for accidents.

```python
# Group the data by weather condition and count the number of accidents for each condition
accidents_by_weather = df.groupby('Weather_Condition').size().reset_index(name='count')

# Set the minimum and maximum counts for scaling the color map
min_count = np.log(accidents_by_weather['count'].min())
max_count = np.log(accidents_by_weather['count'].max())

# Create a color map using Seaborn
cmap = sns.cm.rocket_r

# Create a heatmap using Seaborn
sns.set(rc={'figure.figsize':(12,8)})
sns.set(font_scale=1.5)
sns.heatmap(accidents_by_weather.pivot(index='Weather_Condition', columns='count', values='count'), cmap='plasma', norm=plt.Normalize(vmin=min_count, vmax=max_count), cbar_kws={'ticks': np.arange(min_count, max_count+1), 'label': 'Number of Accidents (log scale)'})
```
